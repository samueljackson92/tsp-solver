```python
%logstart
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tspsolver.tsp_generator import TSPGenerator
from tspsolver.ga.simulator import Simulator
from tspsolver.tuning import GeneticAlgorithmParameterEstimation

OUTPUT_TABLE_DIR = "../report/tables/"
OUTPUT_FIG_DIR = "../report/figures/"
```

Setup the tuner object. This allows us to perform a grid search over all a range of parameters for the GA and choose the best performing one.

```python
NUM_DATASETS = 5 # number of datasets to take the mean fitness over 
NUM_POINTS = 50  # number of points to use in each dataset
tuner = GeneticAlgorithmParameterEstimation(NUM_DATASETS, NUM_POINTS)
```

# Crossover Operators vs. Mutation Operators

First set the parameters for the grid search. Here we are intested in what operators appear to be better and fix the rest of the parameters.

```python
params = {
    "num_epochs": [1000],
    "num_elites": [0],
    "generator": ["SimplePopulationGenerator"],
    "generator_population_size": [20],
    "selector": ["TournamentSelection"],
    "selector_subset_size": [20],
    "selector_tournament_size": [5],
    "crossover": ["OnePointPMX", "TwoPointPMX", "OrderCrossover"],
    "crossover_pcross": [0.9],
    "mutator": ["SwapCityMutation", "DisplacementMutation", "InversionMutation", "InsertionMutation"],
    "mutator_pmutate": [0.1]
}
```

```python
results = tuner.perform_grid_search(params)
```

```python
results
```

```python
cross_vs_mutate = results.pivot(index='crossover', columns='mutator', values='fitness')
cross_vs_mutate
```

```python
cross_vs_mutate.to_latex(OUTPUT_TABLE_DIR + "cross_vs_mutate.tex")
```

Run all parameter sets on a single dataset to show how they converge

```python
gen = TSPGenerator(50)
data = gen.generate()

all_fitness = []
for i, row in results.iterrows():
    params = row.to_dict()
    sim = Simulator(**params)
    sim.evolve(data)
    all_fitness.append(sim.get_min_fitness()[::10])
    
df = pd.DataFrame(np.array(all_fitness))
```

Plot the results of convergence

```python
t = df.T
t.columns = ['-'.join(names) for names in zip(results['crossover'], results['mutator'])]
```

```python
markers = ['+', 'x']

t.plot(style=['-', '--', '-.', ':'] + markers, ylim=(80, 260))
plt.legend(loc='best', bbox_to_anchor=(1.1, 1.05))
plt.savefig(OUTPUT_FIG_DIR + 'cross_vs_mutate_convergence.png', bbox_inches='tight')
```

Run all parameter sets on a single example dataset
